```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
epochs = 20001
```

```python
N = 0
vis_img = mnist.train.images[N:N+1]
```

```python
with tf.Session() as sess:
    new_saver = tf.train.import_meta_graph('save/mnist-model-19999.meta')
    new_saver.restore(sess, 'save/mnist-model-19999')
    
    graph = tf.get_default_graph()
    
    x = graph.get_operation_by_name('input/input-x').outputs[0]
    my_conv1 = graph.get_operation_by_name('conv1/h_conv1').outputs[0]
    conv1_32 = sess.run(my_conv1, feed_dict={x:vis_img})
    conv1_reshape = sess.run(tf.reshape(conv1_32, [32, 1, 28, 28]))
    
    fig, axs= plt.subplots(nrows=2, ncols=16, figsize=(16, 2))
    for num, ax in enumerate(axs):
        for i in range(16):
            ax[i].axis('off')
            ax[i].imshow(conv1_reshape[4*num+i][0])
    plt.title('Conv1 visualization')
    plt.show()
```

```python
with tf.Session() as sess:
    new_saver = tf.train.import_meta_graph('save/mnist-model-'+str(epochs-2)+'.meta')
    new_saver.restore(sess, 'save/mnist-model-'+str(epochs-2))
    
    graph = tf.get_default_graph()
    
    x = graph.get_operation_by_name('input/input-x').outputs[0]
    my_pool1 = graph.get_operation_by_name('conv1/pool/h_pool1').outputs[0]
    pool1_32 = sess.run(my_pool1, feed_dict={x:vis_img})
    pool1_reshape = sess.run(tf.reshape(pool1_32, [32, 1, 14, 14]))
    
    fig, axs= plt.subplots(nrows=2, ncols=16, figsize=(16, 2))
    plt.title('Pool1 visualization')
    for num, ax in enumerate(axs):
        for i in range(16):
            ax[i].axis('off')
            ax[i].imshow(pool1_reshape[4*num+i][0])
    plt.show()
```

```python
with tf.Session() as sess:
    new_saver = tf.train.import_meta_graph('save/mnist-model-'+str(epochs-2)+'.meta')
    new_saver.restore(sess, 'save/mnist-model-'+str(epochs-2))
    
    graph = tf.get_default_graph()
    
    x = graph.get_operation_by_name('input/input-x').outputs[0]
    my_conv2 = graph.get_operation_by_name('conv2/h_conv2').outputs[0]
    conv2_64 = sess.run(my_conv2, feed_dict={x:vis_img})
    conv2_reshape = sess.run(tf.reshape(conv2_64, [64, 1, 14, 14]))
    
    fig, axs= plt.subplots(nrows=4, ncols=16, figsize=(16, 4))
    for num, ax in enumerate(axs):
        for i in range(16):
            ax[i].axis('off')
            ax[i].imshow(conv2_reshape[4*num+i][0])
            
    plt.title('Conv2 visualization')
    plt.show()
```

```python
with tf.Session() as sess:
    new_saver = tf.train.import_meta_graph('save/mnist-model-'+str(epochs-2)+'.meta')
    new_saver.restore(sess, 'save/mnist-model-'+str(epochs-2))
    
    graph = tf.get_default_graph()
    
    x = graph.get_operation_by_name('input/input-x').outputs[0]
    my_pool2 = graph.get_operation_by_name('conv2/pool/h_pool2').outputs[0]
    pool2_64 = sess.run(my_pool2, feed_dict={x:vis_img})
    pool2_reshape = sess.run(tf.reshape(pool2_64, [64, 1, 7, 7]))
    
    fig, axs= plt.subplots(nrows=4, ncols=16, figsize=(16, 4))
    for num, ax in enumerate(axs):
        for i in range(16):
            ax[i].axis('off')
            ax[i].imshow(pool2_reshape[4*num+i][0])
            
    plt.title('Pool2 visualization')
    plt.show()
```
